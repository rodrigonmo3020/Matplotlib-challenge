```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
```

```python
# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single DataFrame
mouse_study_complete = pd.merge(study_results,mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview
mouse_study_complete
```

```python
# Checking the number of mice.
len(mouse_study_complete['Mouse ID'].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.


# mouse_study_complete_a = mouse_study_complete.drop_duplicates(subset=['Mouse ID','Timepoint'], keep='last', inplace=False)
# mouse_study_complete_a
```

```python
# mouse_study_loc = mouse_study_complete_a.loc[mouse_study_complete_a['Mouse ID']== 'g989', :]
# mouse_study_loc
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = mouse_study_complete.loc[mouse_study_complete.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

mouse_study_clean = mouse_study_complete[mouse_study_complete['Mouse ID'].isin(duplicate_mice)==False]
mouse_study_clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(mouse_study_clean['Mouse ID'].unique())
```

# Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean_tumor = mouse_study_clean['Tumor Volume (mm3)'].groupby(mouse_study_clean['Drug Regimen']).mean()
median_tumor = mouse_study_clean['Tumor Volume (mm3)'].groupby(mouse_study_clean['Drug Regimen']).median()
variance_tumor = mouse_study_clean['Tumor Volume (mm3)'].groupby(mouse_study_clean['Drug Regimen']).var()
standard_tumor = mouse_study_clean['Tumor Volume (mm3)'].groupby(mouse_study_clean['Drug Regimen']).std()
SEM_tumor = mouse_study_clean['Tumor Volume (mm3)'].groupby(mouse_study_clean['Drug Regimen']).sem()

summary_stats = pd.DataFrame({'Mean Tumor Volume': mean_tumor,
                             'Median Tumor Volume': median_tumor,
                             'Tumor Volume Variance': variance_tumor,
                             'Tumor Volume Std. Dev.': standard_tumor,
                             'Tumor Volume Std. Err.': SEM_tumor
                             })

summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

summary_stats = mouse_study_clean.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_stats
```

# Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

capomulin_line = capomulin_vol.loc[capomulin_vol["Mouse ID"] == "l509",:]
# capomulin_line.head()
x_axis = capomulin_line["Timepoint"]
tum_size = capomulin_line["Tumor Volume (mm3)"]

fig1, ax1 = plt.subplots()
plt.title('Capomulin treatmeant of mouse l509')
plt.plot(x_axis, tum_size,linewidth=2, markersize=15, color="blue")
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume 
# for the entire Capomulin regimen.

fig1, ax1 = plt.subplots()
avgerage_vol =capomulin_vol.groupby(['Mouse ID']).mean()

marker_size=12
plt.scatter(avgerage_vol['Weight (g)'],avgerage_vol['Tumor Volume (mm3)'], color="blue")
plt.xlabel('Weight (g)',fontsize =11)
plt.ylabel('Averag Tumor Volume (mm3)')

```

# Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(avgerage_vol['Weight (g)'],avgerage_vol['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")
```

```python
(slope, intercept, rvalue, pvalue, stderr)= st.linregress(avgerage_vol["Weight (g)"],avgerage_vol["Tumor Volume (mm3)"])
regress_values=avgerage_vol["Weight (g)"]* slope + intercept

plt.scatter(avgerage_vol["Weight (g)"],avgerage_vol["Tumor Volume (mm3)"],color='b')
plt.plot(avgerage_vol["Weight (g)"], regress_values, color='red')
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Weight vs Tumor Volume for Capomulin")
plt.show()


```
